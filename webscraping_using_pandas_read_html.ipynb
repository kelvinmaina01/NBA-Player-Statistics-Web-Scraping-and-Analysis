## Data Type Conversion and Handling Missing Values

### Subtask:
Convert relevant columns (e.g., 'Age', 'PTS', 'FG%', etc.) to appropriate numeric data types. Identify and decide on a strategy for handling missing values (NaNs) that might be present after web scraping.


**Reasoning**:
First, I will identify the columns that should be numeric and convert them, coercing errors to NaN. Then, I will display the number of missing values per column to understand the extent of missing data before dropping rows with NaNs.



````
numeric_cols = [
    'Rk', 'Age', 'G', 'GS', 'MP', 'FG', 'FGA', 'FG%', '3P', '3PA', '3P%',
    '2P', '2PA', '2P%', 'eFG%', 'FT', 'FTA', 'FT%', 'ORB', 'DRB', 'TRB',
    'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS'
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

print('Number of missing values before dropping rows:')
print(df.isnull().sum())

df.dropna(inplace=True)
print('\nDataFrame shape after dropping rows with NaN values:', df.shape)
````

## Detailed Exploratory Data Analysis (EDA)

### Subtask:
Beyond basic histograms, explore the distributions of other key statistics like rebounds (TRB), assists (AST), and shooting percentages (FG%, 3P%). Identify potential outliers and anomalies in the data.


**Reasoning**:
I will create histograms for 'TRB', 'AST', 'FG%', and '3P%' columns using `seaborn.histplot` to visualize their distributions, as requested by the subtask. I will ensure each plot has relevant labels and titles.



````
import matplotlib.pyplot as plt

# Set up the figure and axes for subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Distribution of Key Player Statistics', fontsize=16)

# Histogram for 'TRB' (Total Rebounds)
sns.histplot(df.TRB, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Distribution of Total Rebounds')
axes[0, 0].set_xlabel('Total Rebounds')
axes[0, 0].set_ylabel('Count')

# Histogram for 'AST' (Assists)
sns.histplot(df.AST, kde=True, ax=axes[0, 1], color='lightcoral')
axes[0, 1].set_title('Distribution of Assists')
axes[0, 1].set_xlabel('Assists')
axes[0, 1].set_ylabel('Count')

# Histogram for 'FG%' (Field Goal Percentage)
sns.histplot(df['FG%'], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title('Distribution of Field Goal Percentage')
axes[1, 0].set_xlabel('Field Goal Percentage')
axes[1, 0].set_ylabel('Count')

# Histogram for '3P%' (3-Point Percentage)
sns.histplot(df['3P%'], kde=True, ax=axes[1, 1], color='mediumpurple')
axes[1, 1].set_title('Distribution of 3-Point Percentage')
axes[1, 1].set_xlabel('3-Point Percentage')
axes[1, 1].set_ylabel('Count')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
````

### Observations from Distributions:

1.  **Total Rebounds (TRB) Distribution:**
    *   The distribution of total rebounds appears right-skewed, with most players having lower rebound numbers and fewer players achieving very high rebound counts. This is typical for many sports statistics, where a few elite players dominate certain categories.
    *   There might be some outliers on the higher end, representing players who are exceptionally good rebounders.

2.  **Assists (AST) Distribution:**
    *   Similar to rebounds, the assists distribution also shows a right-skew, indicating that most players have relatively low assist numbers, while a smaller group of playmakers accumulates significantly more assists.
    *   Potential outliers would be point guards or primary ball-handlers with very high assist averages.

3.  **Field Goal Percentage (FG%) Distribution:**
    *   The field goal percentage distribution seems more spread out and might have a central tendency around the 40-50% mark, which is common in basketball.
    *   There could be players with very low FG% (e.g., role players with limited shot attempts or struggling shooters) and very high FG% (e.g., efficient big men who take high-percentage shots close to the basket).

4.  **3-Point Percentage (3P%) Distribution:**
    *   The 3-point percentage distribution often shows a wider spread than FG%, as 3-point shooting can be more volatile. Many players might have lower 3P% or not attempt many 3-pointers, while specialist shooters will have higher percentages.
    *   Outliers might include players with exceptionally high 3P% on significant volume, or players with very low percentages who still attempt 3-pointers.

## Player Performance Analysis

### Subtask:
Analyze player performance based on various metrics. This could include identifying top players in different categories (e.g., highest points, assists, rebounds), or comparing performance across different player positions.


**Reasoning**:
I will identify and display the top 10 players for 'PTS', 'AST', and 'TRB' by sorting the DataFrame in descending order for each category and selecting the relevant columns.



````
print('Top 10 Players by Points (PTS):')
print(df.sort_values(by='PTS', ascending=False)[['Player', 'PTS']].head(10))

print('\nTop 10 Players by Assists (AST):')
print(df.sort_values(by='AST', ascending=False)[['Player', 'AST']].head(10))

print('\nTop 10 Players by Total Rebounds (TRB):')
print(df.sort_values(by='TRB', ascending=False)[['Player', 'TRB']].head(10))
````

**Reasoning**:
To analyze player performance across different positions, I will calculate the average statistics for each position. This will provide insights into how each position generally performs in key metrics.



````
print('Average statistics per player position:')
print(df.groupby('Pos')[['PTS', 'AST', 'TRB', 'FG%', '3P%', 'STL', 'BLK']].mean())
````

## Team Performance Analysis

### Subtask:
Aggregate individual player statistics to the team level. This will allow for analysis of team-wide performance, identifying top-performing teams, and comparing team statistics.


**Reasoning**:
To analyze team performance, I will aggregate individual player statistics by 'Team' and calculate the mean for the specified columns.



````
team_stats = df.groupby('Team')[['PTS', 'AST', 'TRB', 'FG%', '3P%', 'STL', 'BLK']].mean()
print('Average statistics per team:')
print(team_stats)

print('\nTop 5 Teams by Average Points (PTS):')
print(team_stats.sort_values(by='PTS', ascending=False).head(5))

print('\nTop 5 Teams by Average Assists (AST):')
print(team_stats.sort_values(by='AST', ascending=False).head(5))

print('\nTop 5 Teams by Average Total Rebounds (TRB):')
print(team_stats.sort_values(by='TRB', ascending=False).head(5))
````

## Relationship and Correlation Analysis

### Subtask:
Investigate relationships and correlations between different statistical categories (e.g., the relationship between points and assists, or rebounds and blocks). This can reveal insights into how various aspects of a player's or team's game interact.


**Reasoning**:
I will calculate the correlation matrix for all numeric columns in the DataFrame `df` and then print specific correlation coefficients to investigate relationships between different statistical categories as requested.



````
correlations = df.corr(numeric_only=True)

print('Correlation between PTS and AST:', correlations.loc['PTS', 'AST'])
print('Correlation between PTS and TRB:', correlations.loc['PTS', 'TRB'])
print('Correlation between AST and TRB:', correlations.loc['AST', 'TRB'])
print('Correlation between FG% and PTS:', correlations.loc['FG%', 'PTS'])
print('Correlation between STL and BLK:', correlations.loc['STL', 'BLK'])
````

**Reasoning**:
To visualize the relationships between all numerical features, I will create a heatmap of the correlation matrix, which will provide a comprehensive overview of how each statistical category correlates with others.



````
plt.figure(figsize=(16, 12))
sns.heatmap(correlations, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix of NBA Player Statistics', fontsize=18)
plt.show()
````
